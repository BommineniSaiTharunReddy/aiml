```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('/content/titanic.csv')
df.head()
```

```python
# Plot numeric features
sns.pairplot(df[['Age', 'Fare', 'SibSp', 'Parch']])
plt.show()
```

```python
# Relative size of survived/unsurvived
sns.countplot(x='Survived', data=df)
plt.xlabel("Survival Status (0: Not Survived, 1: Survived)")
plt.show()
```

```python
# Check patterns on gender
sns.countplot(x='Sex', data=df, hue='Survived')
plt.xlabel("Gender")
plt.show()
```

```python
# Passenger class-wise survival rate
sns.barplot(x='Pclass', y='Survived', data=df)
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```

```python
# Siblings and Spouses distribution
sns.countplot(x='SibSp', data=df, hue='Survived')
plt.xlabel("Number of Siblings/Spouses Aboard")
plt.show()
```

```python
# Overall age distribution
sns.histplot(df['Age'].dropna(), kde=True)
plt.xlabel("Age")
plt.show()
```

```python
# Class-wise age distribution
sns.boxplot(x='Pclass', y='Age', data=df)
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()
```
